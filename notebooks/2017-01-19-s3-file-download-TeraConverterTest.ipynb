```python
af.download_s3("76AO/Strips/000000.dcimg", "../data/000000.dcimg")

s3 = boto3.resource('s3')

bucket = s3.Bucket('alpenglowoptics')

ll = list(bucket.objects.all())

fname_list = []
for l in ll:
    k = l.key
    if k.startswith('76AO/Strips/im') and not k.endswith('_max.tif'):
        fname_list.append(k)
```

```python
fname_list[-1].split("_")
```

```python
import os.path as op
```

```python
op.split(fname_list[-1])
```

```python
numbers = [int(x) for x in op.split(fname_list[-1])[-1].split(".")[0].split("_")[1:]]
numbers.insert(0,1)
numbers
```

```python
# cleaning up directories (ignore)
for chan in range(numbers[0]):
    for strip in range(numbers[1]):
        os.rmdir("%06d_%06d"%(chan, strip))
```

```python
# how to automate making directory structure
for chan in range(numbers[0]):
    for strip in range(numbers[1]):
        os.makedirs("%06d/%06d"%(chan, strip))
```

```python
for f in fname_list:
    chan = 0
    strip = int(f.split("_")[1])
    af.download_s3(f, op.join("./%06d/%06d"%(chan,strip), op.split(f)[-1]))
```
